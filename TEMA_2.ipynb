# **ESTADÍSTICA**
La estadística es una rama de las matemáticas que se encarga de recolectar, organizar, analizar, interpretar y presentar datos. Su objetivo principal es ayudar a tomar decisiones informadas a partir de datos y a comprender mejor fenómenos complejos. La estadística se divide en dos áreas principales:

*   **Estadística descriptiva**: Se enfoca en describir y resumir los datos. Esta parte de la estadística no infiere ni hace predicciones; simplemente describe lo que muestran los datos.
*  **Estadística inferencial**: Utiliza una muestra de datos para hacer inferencias o predicciones sobre una población más amplia. Incluye técnicas como la estimación de parámetros, pruebas de hipótesis, regresión, y análisis de varianza (ANOVA). Este tipo de estadística ayuda a hacer generalizaciones sobre una población a partir de los datos de una muestra.


Una **variable aleatoria** es una función que asigna un valor numérico a cada resultado posible de un experimento aleatorio. En términos simples, es una manera de cuantificar los resultados inciertos de un evento, donde los valores que puede tomar están sujetos al azar. Las variables aleatorias pueden ser **discretas**, tomando un número finito de valores, o **continuas**, tomando cualquier valor dentro de un rango.

### Variable Aleatoria de Bernoulli

Una **variable aleatoria de Bernoulli o binaria** es el ejemplo más sencillo de variable aleatoria discreta. Este tipo de varibles toma los valores 1 (éxito) o 0 (no éxito) y sólo se requiere conocer un parámetros para describir su comportamiento, la probabilidad de que la variable tome el valor 1.

Por ejemplo, pensemos en el lanzamiento de una moneda equilibrada. Si a sacar cara le damos el valor 1 y cruz el valor 0, entonces $p(X=1)=0,5$. Como la suma de las probabilidades de todos los valores de una variable aleatoria deben sumar uno se deduce que $p(X=0)=0,5$.<br><br>

En el ámbito financiero, las decisiones de inversión se toman en un entorno de incertidumbre. Los rendimientos de las inversiones, los precios de los activos, y otros factores económicos clave se comportan de manera incierta, lo que significa que no podemos predecir sus valores futuros con certeza. Aquí es donde entran en juego las variables aleatorias.

Empezaremos viendo las principales herramientas para describir los rendimientos de los activos.

### Variable Aleatoria Discreta

De manera más general, cualquier **variable aleatoria discreta** se define por el conjunto de valores que puede tomar así como las probabilidades de que tome cada uno de ellos. Si los valores x que puede tomar la variable $X$ son ${x_1,x_2…,x_n}$, entonces las probabilidades $p_1, p_2, …,p_n$ se definen mediante

$$p_i=P(X=x_i)\qquad i=1, 2, …, n$$

donde:
- $p_i$ es la probabilidad de que la variable aleatoria $X$ tome el valor $x_i$

## Propiedades de la Probabilidad
- Para cualquier evento $x_i$, $P(x_i) \geq 0$.
- La probabilidad del espacio muestral completo es 1, es decir, $P(S) = 1$.
-cada $p_i$ está entre 0 y 1
- La **probabilidad de la suma** se refiere a la probabilidad de que ocurra al menos uno de dos o más eventos. Por ejemplo, si $A$ y $B$ son dos eventos,  la probabilidad de la suma es la probabilidad de que ocurra $A$, $B$, o ambos. Esta probabilidad se denota como $P(A \cup B)$ , y se calcula usando la siguiente fórmula:
$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$
El término $P(A \cap B)$ se resta para evitar contar dos veces la probabilidad de la intersección de $A$ y $B$, es decir, la probabilidad de que ambos eventos ocurran simultáneamente.

  Si los eventos $A$ y $B$ son mutuamente excluyentes (no pueden ocurrir al mismo tiempo), entonces $P(A \cap B) = 0$, y la fórmula se simplifica a:

$$
P(A \cup B) = P(A) + P(B)
$$

**Nota: Definición Frecuentista de Probabilidad:**
$$\text{Probabilidad}= \frac{\text{Casos Favorables}}{\text{Casos Posibles}}$$

### Ejemplo

Considera dos eventos:
- $A$: Sacar un número impar al lanzar un dado  (probabilidad $P(A) = \frac{3}{6} = \frac{1}{2}$).
- $B$: Sacar un número menor o igual a 3  (probabilidad $P(B) = \frac{3}{6} = \frac{1}{2}$).

Primero, calculemos $P(A \cap B)$. Ambos eventos ocurren si sacamos un 1 o un 3, así que:

$$
P(A \cap B) = \frac{2}{6} = \frac{1}{3}
$$

Ahora, la probabilidad de que ocurra $A$ o $B$ (o ambos) es:

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$<Br><Br>
$$
P(A \cup B) = \frac{1}{2} + \frac{1}{2} - \frac{1}{3} = 1 - \frac{1}{3} = \frac{2}{3}
$$

- **Probabilidad Conjunta** cuantifica la probabilidad de que dos o más eventos ocurran simultáneamente. Si $A$ y $B$ son dos eventos, la probabilidad conjunta se denota como $P(A \cap B)$ o $P(A \text{ y } B)$.

- **Probabilidad Condicionada**:  La probabilidad de un evento $A$ dado que otro evento $B$ ha ocurrido se llama probabilidad condicional y se denota $P(A|B)$. Se calcula como

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$
  siempre que $P(B) > 0$.

- **Eventos Independientes**: Dos eventos son independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. $P(B|A) = P(B)$ , $P(A|B) = P(A)$.


  Para eventos dependientes, la probabilidad conjunta se calcula usando la probabilidad condicional:
$$
P(A \cap B) = P(A) \cdot P(B|A)
$$
  Aquí, $P(B|A)$ es la probabilidad de que ocurra $B$ dado que $A$ ya ha ocurrido.

  Para eventos independientes, la probabilidad conjunta se simplifica, ya que $P(B|A) = P(B)$:
$$
P(A \cap B) = P(A) \cdot P(B)
$$

### Ejemplo

Supongamos que lanzamos dos veces un dado y definimos los siguientes eventos:
- $A$: Sacar un 3 en el primer lanzamiento (probabilidad $P(A) = \frac{1}{6}$).
- $B$: Sacar un número par al lanzar el mismo dado la segunda vez (probabilidad $P(B) = \frac{3}{6} = \frac{1}{2}$).

Si los eventos son independientes, la probabilidad conjunta de sacar un 3 y un número par sería:

$$
P(A \cap B) = P(A) \cdot P(B) = \frac{1}{6} \cdot \frac{1}{2} = \frac{1}{12}
$$


- **Teorema de Bayes**
Es una fórmula fundamental que relaciona las probabilidades condicionales y marginales de dos eventos. El Teorema de Bayes nos permite actualizar nuestra creencia sobre la probabilidad de un evento en función de nueva evidencia. Inicialmente, tenemos una idea ("a priori") de la probabilidad de que ocurra un evento \( A \). Luego, cuando observamos un nuevo dato o evidencia \( B \), el teorema nos proporciona una forma de actualizar esta creencia a una nueva probabilidad ("posterior").

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$
  Donde:

- $P(A|B)$: Es la probabilidad asociada al evento \( A \) dado que ya ha ocurrido el evento \( B \). Se llama la **probabilidad posterior**.
- $P(A)$: Es la probabilidad asociada al evento \( A \) sin ninguna información sobre \( B \). Se llama la **probabilidad a priori** o  **prior**.<br><br>

  Un ejemplo clásico es el diagnóstico médico y los cribados. Imagina que estás usando una prueba para detectar una enfermedad rara. Supongamos:
- $P(A)$ es la probabilidad de tener la enfermedad en la población (por ejemplo, 1%).
- $P(B|A)$ es la probabilidad de que la prueba sea positiva si tienes la enfermedad (por ejemplo, 99%).
- $P(B)$ es la probabilidad de que la prueba sea positiva, sin importar si tienes o no la enfermedad. Por ejemplo un 2%.

  El Teorema de Bayes te permite calcular la probabilidad de que realmente tengas la enfermedad, dado que obtuviste un resultado positivo en la prueba:
  $$P(\text{Enfermedad}|\text{Prueba Positiva}) = \frac{P(\text{Prueba Positiva}|\text{Enfermedad}) \cdot P(\text{Enfermedad})}{P(\text{Prueba Positiva})}$$<Br><Br>
  $$P(\text{Enfermedad}|\text{Prueba Positiva}) = \frac{99\% \cdot 1\%}{2\%}=49,5\%$$

#REPRESENTACIÓN DE VARIABLES ALEATORIAS

La **función de densidad de probabilidad**(abreviada como f.d.p. o pdf por sus siglas en inglés, probability density function) de $X$ resume la información de los valores que puede tomar $X$ y a sus correspondientes probabilidades:

$$f(x_i)=p_i\qquad i=1, 2, …, n$$


```python
# @title
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
%matplotlib inline
from IPython.core.display import HTML
def centergraph():
    display(HTML("""
    <style>
    #output-body {
        display: flex;
        align-items: center;
        justify-content: center;
    }
    </style>
    """))
    return ""
centergraph()
# Datos de ejemplo: lanzamiento de un dado
frequencies = [1/6] * 6  # Cada cara del dado tiene la misma probabilidad
faces = np.arange(1, 7)
# Crear el gráfico de barras para la distribución de frecuencias
plt.bar(faces, frequencies, color='blue', edgecolor='black')

# Añadir títulos y etiquetas
plt.title('Lanzamiento de un Dado')
plt.xlabel('x=Cara del Dado')
plt.ylabel('f(x)')
# Mostrar el gráfico
plt.show()
```



Para las variables aleatorias continuas también se puede definir una función de densidad de probabilidad para describirla, sin embargo, dado que la probabilidad de que una variable aleatoria continua tome un determinado valor entre los infinitos que puede tomar es 0, en estos casos la **fdp** se usa para calcular eventos que comprenden un rango de valores.
La probabilidad de que la variable aleatoria $X$ tome un valor dentro de un intervalo $[a, b]$ se calcula como el área bajo la curva de $f(x)$ entre $a$ y $b$:
$$P(a \leq X \leq b)=\int_{a}^{b}f(x)\, dx$$

- **No Negatividad**:
   - La f.d.p. es siempre no negativa para todos los valores de $x$:
   $$f(x) \geq 0 \quad \text{para todo } x$$

- **Área Bajo la Curva Igual a 1**:
   - La integral de la f.d.p. sobre todo el espacio de valores posibles debe ser igual a 1, lo que asegura que la suma de las probabilidades para todos los valores posibles es 1:
   $$\int_{-\infty}^{\infty} f(x) \, dx = 1$$



```python

# @title
from scipy.stats import norm

# Definir los parámetros de la distribución normal estándar
mean = 0
std_dev = 1
# Generar valores x
x = np.linspace(-4, 4, 1000)
# Calcular la función de densidad de probabilidad
y = norm.pdf(x, mean, std_dev)
# Crear la figura y el eje
fig, ax = plt.subplots()

# Dibujar la curva de la función de densidad
ax.plot(x, y, label='Distribución Normal 0-1', color='blue')
# Rellenar el área bajo la curva entre -1 y 1
x_fill = np.linspace(-1, 1, 1000)
y_fill = norm.pdf(x_fill, mean, std_dev)
ax.fill_between(x_fill, y_fill, color='green', alpha=0.5, label='Área entre -1 y 1')

# Añadir títulos y etiquetas
ax.set_title('Función de Densidad')
ax.set_xlabel('Valores')
ax.set_ylabel('Densidad de Probabilidad')
ax.text(0.8, 0.2, 'p(-1≤x≤1)', fontsize=12, color='black', ha='right')

# Mostrar el gráfico
plt.show()
centergraph()
```

Para calcular probabilidades de variables aleatorias continuas, es útil el uso de la **función de distribución acumulada** (fda). Si $X$ es cualquier variable aleatoria, entonces para cualquier número real $x$ su fda está definida por
 $$F(x)=P(X\leq x)$$

En el caso de variables aleatorias discretas, sería la suma de las fdp de todos los valores $x_i$ tales que $x_i\leq x$
En el caso de variables aleatorias continuas, $F(x)$ es el área bajo la fdp, $f$, a la izquierda del punto $x$.<br><br>


```python
# @title
# Dibujar la curva de la función de densidad
fig, ax = plt.subplots()
ax.plot(x, y, label='Distribución Normal 0-1', color='blue')

# Rellenar el área bajo la curva para x <= 1
x_fill = np.linspace(-4, 1, 1000)
y_fill = norm.pdf(x_fill, mean, std_dev)
ax.fill_between(x_fill, y_fill, color='green', alpha=0.5, label='Área x ≤ 1')

# Añadir títulos y etiquetas
ax.set_title('Función de Densidad')
ax.set_xlabel('Valores')
ax.set_ylabel('Densidad de Probabilidad')
ax.text(0.8, 0.2, 'F(x)=p(x≤1)', fontsize=12, color='black', ha='right')

# Mostrar el gráfico
plt.show()
centergraph()
```

## **REPRESENTACIONES GRÁFICAS**

**Tablas de frecuencias**: Organizan los datos en categorías o intervalos, mostrando cuántas veces aparece cada valor o rango de valores.


*   Frecuencia Absoluta: número de veces que aparece
*   Frecuencia Relativa: % de veces que aparece

En el caso de variables ordenadas de cada una de estas frecuencias podemos ver los datos acumulados, es decir la frecuencia acumulada de todas las categorías inferiores o iguales a la de interés.


```python
# @title
data['Intervalo']=pd.cut(np.array(data['Ret']), bins=10)

# Calcular la frecuencia absoluta
tabla_frecuencias_abs = pd.crosstab(index=data['Intervalo'], columns='Frecuencia Absoluta')


# Calcular la frecuencia relativa
tabla_frecuencias_rel = pd.crosstab(index=data['Intervalo'], columns='Frecuencia Relativa',normalize=True)*100
tabla_frecuencia_combinada = pd.concat([tabla_frecuencias_abs, tabla_frecuencias_rel], axis=1)
# Calcular la frecuencia absoluta acumulada
tabla_frecuencia_combinada['Frecuencia Absoluta Acumulada']= tabla_frecuencias_abs.cumsum()
tabla_frecuencia_combinada['Frecuencia Relativa Acumulada']= tabla_frecuencias_rel.cumsum()
tabla_frecuencia_combinada


```


## **MEDIDAS DE TENDENCIA CENTRAL**

Las medidas de tendencia central son estadísticos que describen el valor central o típico de un conjunto de datos. Son útiles porque proporcionan un punto de referencia para comprender y comparar datos y entender la distribución de un conjunto de valores. Las principales medidas de tendencia central son la **media**, la **mediana** y la **moda**.


### Esperanza Matemática / Media

La **esperanza matemática** (también conocida como valor esperado) es un concepto fundamental en probabilidad y estadística, y juega un papel crucial en el análisis financiero. En términos simples, la esperanza matemática es el promedio ponderado de todos los posibles valores que puede tomar una variable aleatoria, donde los pesos son las probabilidades de cada uno de esos valores.


Para una variable aleatoria discreta $X$ que toma valores $x_1, x_2, ..., x_n$ con probabilidades asociadas $p_1, p_2, ..., p_n$, la esperanza matemática $E(X)$ se define como:
$$E(X) = \sum_{i=1}^{n} x_i \cdot p_i$$

Para una variable aleatoria continua, la esperanza matemática se calcula integrando la función de densidad de probabilidad $f(x)$:
$$E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx$$

Algunas propiedades de la espranza matemática/media:

Propiedad de Linealidad:

- La media de la suma de dos o más variables aleatorias es igual a la suma de las medias de esas variables.
Matemáticamente: $E(aX + bY) = aE(X) + bE(Y)$, donde $a$ y $b$ son constantes.
- Si a cada observación de un conjunto de datos se le suma una constante $c$, la media del nuevo conjunto de datos será la media original más esa constante.
Matemáticamente: Si $Y_i = X_i + c$, entonces $E(Y) = E(X) + c$.
- Si cada observación de un conjunto de datos se multiplica por una constante $c$, la media del nuevo conjunto será $c$ veces la media original.
Matemáticamente: Si $Y_i = cX_i$, entonces $E(Y) = cE(X)$.

La esperanza matemática es esencial para evaluar y comparar diferentes activos, ya que permite a los inversores calcular el rendimiento promedio esperado de una inversión a lo largo del tiempo. Por ejemplo: Si una acción tiene un 50% de probabilidad de generar un rendimiento del 10% y un 50% de probabilidad de generar un rendimiento del 5%, el rendimiento esperado sería:

   $$
   E(\text{Rendimiento}) = 0.5 \times 10\% + 0.5 \times 5\% = 7.5\%
   $$

En el mundo real lo que observamos para cada activos son las observaciones de los rendimientos, no conocemos la distribución real que las genera. Por eso hacemos uso de **estimadores** como la media.


### Media (Promedio)

La **media** es el promedio aritmético de un conjunto de datos. Se calcula sumando todos los valores y luego dividiendo por el número total de observaciones.

- **Fórmula**:

La fórmula para calcular la media es:

$$ \text{Media} = \frac{\sum_{i=1}^{n} x_i}{n} $$


  donde $x_i$ representa cada valor en el conjunto de datos y $n$ es el número total de observaciones.

- **Ventajas**:
  - Fácil de calcular e interpretar.
  - Utiliza todos los datos del conjunto, lo que la hace sensible a todos los valores.

- **Desventajas**:
  - Es sensible a los valores extremos (outliers), que pueden distorsionar la media.

La media que podemos calcular con los datos que observamos en la realidad sirven para estimar cuál es el valor de la esperanza matemática de los rendimientos del activo.

Otras referencias que se usan al analizar los rendimientos de un activos son:<br><br>

### Mediana

La **mediana** es el valor que divide un conjunto de datos ordenados en dos partes iguales. Es el punto medio cuando los datos están organizados de menor a mayor.

- **Cálculo**:
  - Si el número de observaciones es impar, la mediana es el valor central.
  - Si el número de observaciones es par, la mediana es el promedio de los dos valores centrales.

- **Ventajas**:
  - No se ve afectada por valores extremos, lo que la hace útil cuando el conjunto de datos tiene outliers.

- **Desventajas**:
  - No utiliza toda la información del conjunto de datos (solo se enfoca en el valor central).<br><br>

### Moda

La **moda** es el valor o valores que aparecen con mayor frecuencia en un conjunto de datos. Un conjunto de datos puede tener una moda, más de una (bimodal, multimodal) o no tener ninguna moda si todos los valores son únicos.

- **Ventajas**:
  - Es útil para datos categóricos, donde no tiene sentido calcular la media o la mediana.
  - Indica el valor más común o típico en un conjunto de datos.

- **Desventajas**:
  - En conjuntos de datos con muchas observaciones únicas, la moda puede no ser representativa.
  - No siempre existe una moda única, lo que puede complicar su interpretación.

## Ejemplos Prácticos

- **Media**: Si tienes los valores [3, 5, 7, 9], la media es $\frac{3+5+7+9}{4} = 6$.
  
- **Mediana**: Para los valores [3, 5, 7, 9], la mediana es 6. Si tuvieras $[3, 5, 7]$, la mediana sería 5.

- **Moda**: Para los valores [3, 5, 7, 5, 9], la moda es 5, ya que es el valor que aparece con más frecuencia.


```python
media = sum(data['Ret'].dropna()) / len(data['Ret'].dropna())
print(media)
# Otra forma
media = data['Ret'].dropna().mean()
print(media)
# Otra forma
mediana=data['Ret'].dropna().median()
# Combinar texto y valor calculado usando f-string
print(f"La media de los valores es: {media:.3f}%")
print(f"La mediana de los valores es: {mediana:.3f}%")
```


Además de las medidas de tendencia central tenemos también otras medidas de localización útiles como son los Percentiles y Cuartiles.

Los percentiles, cuartiles, deciles son medidas estadísticas utilizadas para describir la posición de un valor dentro de un conjunto de datos, especialmente en distribuciones que no son simétricas. Estas medidas dividen un conjunto de datos en partes y permiten entender la distribución de los valores.

Son especialmente usadas en el ámbito de los riesgos financieros pues permiten conocer cómo es la distribución de los rendimientos cuando estos son muy negativos

### Percentiles
Un **percentil** es un valor en un conjunto de datos por debajo del cual se encuentra un cierto porcentaje de los datos. Por ejemplo, el percentil 25 es el valor por debajo del cual se encuentra el 25% de los datos.

Si un conjunto de datos está ordenado de menor a mayor, el percentil $p$ se encuentra en la posición $i$ calculada como:
  $$ i = \frac{p}{100} \times (n + 1)$$
  donde:
  - $p$ es el percentil que deseas calcular (por ejemplo, 25 para el percentil 25).
  - $n$ es el número total de datos.
  - $i$ es la posición del percentil dentro del conjunto de datos.

Ejemplo:
- Si tienes un conjunto de datos con 100 valores, el percentil 90 es el valor por debajo del cual se encuentran el 90% de los datos.

### Cuartiles

Los **cuartiles** son un caso especial de percentiles que dividen un conjunto de datos en cuatro partes iguales. Cada cuartil representa un 25% del conjunto de datos.

Ejemplos:
1. Primer Cuartil (Q1):
   - También conocido como el **percentil 25**.
   - El 25% de los datos están por debajo de este valor.

2. Segundo Cuartil (Q2):
   - Es el **percentil 50**, que es también la **mediana** del conjunto de datos.
   - El 50% de los datos están por debajo de este valor.

3. Tercer Cuartil (Q3):
   - También conocido como el **percentil 75**.
   - El 75% de los datos están por debajo de este valor.

Ejemplo:

  - **Q1** sería 2.5 (entre 2 y 3).
  - **Q2** (mediana) sería 5.
  - **Q3** sería 7.5 (entre 7 y 8).

Estas medidas son esenciales para el análisis estadístico, permitiendo una comprensión profunda de cómo se distribuyen los datos en un conjunto.

# Medidas Estadísticas de Dispersión

Las medidas estadísticas de dispersión son métricas que indican cuán dispersos o extendidos están los valores en un conjunto de datos. Mientras que las medidas de tendencia central (como la media, la mediana y la moda) nos proporcionan información sobre el valor central o típico de los datos, las medidas de dispersión nos indican la variabilidad o el grado de dispersión de esos datos alrededor del valor central.

## Principales Medidas de Dispersión

### Rango (o Amplitud)

- **Definición**: El rango es la diferencia entre el valor máximo y el valor mínimo en un conjunto de datos.
- **Fórmula**:
  $$\text{Rango} = \text{Valor máximo} - \text{Valor mínimo}$$
- **Ventajas**:
  - Es fácil de calcular.
- **Desventajas**:
  - Es muy sensible a los valores extremos (outliers), ya que solo depende de los dos valores más extremos.
  - No proporciona información sobre la dispersión de los datos entre los valores extremos.


### Rango Intercuartílico (RIC)

- **Definición**: El rango intercuartílico es la diferencia entre el tercer cuartil (Q3) y el primer cuartil (Q1) de un conjunto de datos. Representa la dispersión de la mitad central de los datos.
- **Fórmula**:
  $$\text{RIC} = Q3 - Q1$$
- **Ventajas**:
  - No se ve afectado por los valores extremos, lo que lo hace útil en la presencia de outliers.
  - Proporciona una medida de dispersión centrada en la mediana.
- **Desventajas**:
  - Ignora los datos fuera del rango intercuartílico (es decir, los cuartiles extremos).

### Varianza

La **varianza** de una variable aleatoria es una medida de dispersión que indica qué tan lejos, en promedio, se encuentran los valores de la variable respecto a su media. En otras palabras, la varianza nos dice cuánto varían o se dispersan los posibles valores de una variable aleatoria alrededor de su valor esperado (media).


Para una variable aleatoria $X$, con esperanza matemática $E(X) = \mu$, la varianza $\sigma^2 $ se define como el valor esperado del cuadrado de las desviaciones de $X$ respecto a su media:

$$\sigma^2 = \text{Var}(X) = E\left[(X - \mu)^2\right]$$

Esto puede interpretarse como el promedio ponderado de las diferencias al cuadrado entre los valores de $X$ y su media, donde las probabilidades de los valores de $X$ sirven como pesos.

### Cálculo de la Varianza:

- **Para una Variable Aleatoria Discreta**:
  Si $X$ es una variable aleatoria discreta que toma los valores $x_1, x_2, \dots, x_n$ con probabilidades asociadas $p_1, p_2, \dots, p_n$, la varianza se calcula como:
  
  $$\sigma^2 = \sum_{i=1}^{n} p_i \cdot (x_i - \mu)^2$$
  Donde $ \mu = E(X) = \sum_{i=1}^{n} p_i \cdot x_i $.

- **Para una Variable Aleatoria Continua**:
  Si $X$ es una variable aleatoria continua con función de densidad de probabilidad $f(x)$, la varianza se calcula como:
  
  $$
  \sigma^2 = \int_{-\infty}^{\infty} (x - \mu)^2 \cdot f(x) \, dx
  $$

### Ejemplo:

Supongamos que tienes una variable aleatoria discreta $X$ que representa el resultado de lanzar un dado equilibrado, con los valores posibles $1, 2, 3, 4, 5, 6$ y probabilidades iguales de $\frac{1}{6}$. Calcular la Varianza:

   $
   \text{Var}(X) = \sum_{i=1}^{6} \frac{1}{6} \cdot (x_i - 3.5)^2
    = \frac{1}{6}[(1 - 3.5)^2 + (2 - 3.5)^2 + (3 - 3.5)^2 + (4 - 3.5)^2 + (5 - 3.5)^2 + (6 - 3.5)^2]
   = \frac{1}{6}[(2.5)^2 + (1.5)^2 + (0.5)^2 + (-0.5)^2 + (-1.5)^2
   = \frac{1}{6}(6.25 + 2.25 + 0.25 + 0.25 + 2.25 + 6.25) = \frac{1}{6}(17.5) = 2.92
   $
La varianza de $X$ es aproximadamente 2.92.<br><br>
En el análisis financiero, la varianza es una medida clave del riesgo de una inversión. A mayor varianza mayor posibilidad de pérdidas (y beneficios).


```python
# @title
# Generar observaciones aleatorias
np.random.seed(42)
n_observaciones = 1000

# Distribución normal 1
media1=0.003
desves1=0.005
dist_1 = np.random.normal(loc=media1, scale=desves1, size=n_observaciones)

media2=0.001
desves2=0.01
# Distribución normal 2
dist_2 = np.random.normal(loc=media2, scale=desves2, size=n_observaciones)

# Determinar los límites comunes para el eje y
y_min = min(dist_1.min(), dist_2.min())
y_max = max(dist_1.max(), dist_2.max())

# Crear gráficos de dispersión directamente con las distribuciones normales
plt.figure(figsize=(14, 6))

# Gráfico 1
plt.subplot(1, 2, 1)
plt.scatter(range(n_observaciones), dist_2, alpha=0.6, color='orange')
plt.axhline(media1, color='orange', linestyle='--', linewidth=2)
plt.title('Activo A')
plt.xlabel('Observaciones')
plt.ylabel('Rendimientos')
plt.ylim(y_min, y_max)  # Fijar la escala del eje y
plt.grid(True)
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x*100:.2f}%'))
# Gráfico 2
plt.subplot(1, 2, 2)
plt.scatter(range(n_observaciones), dist_1, alpha=0.6)
plt.title('Activo B')
plt.xlabel('Observaciones')
plt.ylabel('Rendimientos')
plt.ylim(y_min, y_max)  # Fijar la escala del eje y
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x*100:.2f}%'))
plt.axhline(media2, linestyle='--', linewidth=2)
plt.grid(True)

# Mostrar los gráficos
plt.tight_layout()
plt.show()
centergraph()
```


Al igual que en el caso de media estimamos la varianza a traves de los datos reales. El estadístico que usamos es:

Para una población completa o con muchos datos:
  $$\sigma^2 = \frac{\sum_{i=1}^{N} (x_i - \mu)^2}{N}$$
Para una muestra:
 $$\sigma^2 = \frac{\sum_{i=1}^{N} (x_i - \mu)^2}{N-1}$$
  Donde:
  - $\sigma^2 $ es la varianza.
  - $ x_i $ son los valores del conjunto de datos.
  - $ \mu $ es la media del conjunto de datos.
  - $ N $ es el número de observaciones.
- **Ventajas**:
  - Usa todos los datos del conjunto, proporcionando una medida completa de la dispersión.
- **Desventajas**:
  - Tiene unidades al cuadrado, lo que puede ser difícil de interpretar en comparación con los datos originales.
- **Algunas Propiedades**:
  - Si a todos los datos de un conjunto se les suma o resta una constante, la varianza no cambia. Es decir, si $Y_i = X_i + c$, entonces $\sigma_Y^2 = \sigma_X^2$.
  - Si todos los datos de un conjunto se multiplican por una constante, la varianza también se multiplica por el cuadrado del valor de esa constante. Matemáticamente, si $Y_i = cX_i$, entonces $\sigma_Y = c^2\sigma_X$.

### **Desviación Estándar**

- **Definición**: La desviación estándar es la raíz cuadrada de la varianza. Proporciona una medida de dispersión en las mismas unidades que los datos originales. Cuando se habla de activos financieros a la desviación estándar de los retornos se le suele denominar ***volatilidad***.
 $$  \sigma =\sqrt{\sigma^2}$$

- **Fórmula del estimador**:
En la población:
  $$
  \sigma = \sqrt{\frac{\sum_{i=1}^{N} (x_i - \mu)^2}{N}}
  $$
- **Ventajas**:
  - Es fácil de interpretar porque está en las mismas unidades que los datos originales.
- **Desventajas**:
  - Como la varianza, es sensible a los valores extremos.
- **Algunas Propiedades**:
  - Si a todos los datos de un conjunto se les suma o resta una constante, la varianza no cambia. Es decir, si $Y_i = X_i + c$, entonces $\sigma_Y = \sigma_X$.
  - Si todos los datos de un conjunto se multiplican por una constante, la desviación típica también se multiplica por el valor absoluto de esa constante. Matemáticamente, si $Y_i = cX_i$, entonces $\sigma_Y = |c|\sigma_X$.


### Coeficiente de Variación

- **Definición**: El coeficiente de variación es una medida relativa de dispersión que se obtiene dividiendo la desviación estándar por la media y a menudo se expresa como un porcentaje.
- **Fórmula**:
  $$\text{CV} = \frac{\sigma}{\mu} \times 100$$
- **Ventajas**:
  - Permite comparar la dispersión entre diferentes conjuntos de datos, incluso si las unidades de medida son diferentes.
- **Desventajas**:
  - No es adecuado cuando la media es cercana a cero, ya que puede producir valores engañosos.

  Calculemos estas medidas en nuestro ejemplo:

```python
varianza=data['Ret'].var(ddof=0)
desvest=data['Ret'].std(ddof=0)
RangInt=percentiles[0.75]-percentiles[0.25]
RangMax=data['Ret'].max()-data['Ret'].min()
cv=desvest/media

print(f"La varianza: {varianza:.2f}")
print(f"La desviación típica: {desvest:.2f}%")
print(f"El rango intercuantílico es:{RangInt:.2f}%")
print(f"El rango: {RangMax:.2f}%")
print(f"El coeficiente de variación es: {cv:.2f}")
```

##**Estandarización de Variables**
Estandarizar una variable es el proceso de transformar una variable para que tenga una media de 0 y una desviación típica de 1. Este procedimiento es comúnmente utilizado para comparar variables que originalmente tienen diferentes unidades de medida o escalas.

La estandarización se realiza mediante la siguiente fórmula:
$$z=\frac{x-\mu}{\sigma}$$

donde:

$X$ es el valor original de la variable.
$\mu$ es la media de la variable.
$\sigma$ es la desviación típica de la variable.
El resultado, $Z$, es conocido como la variable estandarizada o puntaje Z.

Al estandarizar las variables, se eliminan las diferencias de escala y se pueden comparar diferentes variables en términos de desviaciones estándar.


## **Asimetría**

### **Definición:**
- La asimetría, también conocida como **skewness** en inglés, mide la simetría de una distribución de datos. Indica si los datos están distribuidos de manera uniforme alrededor de la media o si hay una inclinación hacia un lado u otro.

### **Tipos de Asimetría:**

1. **Asimetría Positiva (Sesgo a la Derecha)**:
   - La cola de la distribución es más larga hacia la derecha.
   - La mayoría de los valores se concentran en la parte baja de la escala, y hay algunos valores extremadamente altos que arrastran la media hacia la derecha.

2. **Asimetría Negativa (Sesgo a la Izquierda)**:
   - La cola de la distribución es más larga hacia la izquierda.
   - La mayoría de los valores se concentran en la parte alta de la escala, y hay algunos valores extremadamente bajos que arrastran la media hacia la izquierda.

3. **Distribución Simétrica**:
   - La asimetría es aproximadamente 0.
   - La distribución tiene colas simétricas a ambos lados de la media.
   - Ejemplo: La distribución normal (campana de Gauss).

### **Cálculo:**
- La fórmula general para calcular la asimetría es:

  $$\text{Asimetría} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \bar{x}}{\sigma}\right)^3$$
  donde:
  - $ n $ es el número de observaciones.
  - $ x_i $ es cada valor del conjunto de datos.
  - $ \bar{x} $ es la media.
  - $ s $ es la desviación estándar.

### **Interpretación:**
- **Asimetría positiva**: Valor positivo indica que la cola derecha es más larga o gruesa que la izquierda.
- **Asimetría negativa**: Valor negativo indica que la cola izquierda es más larga o gruesa que la derecha.
- **Asimetría cercana a 0**: Indica una distribución aproximadamente simétrica.

En términos de retornos, una asimetría negativa por ejemplo quiere decir que es frecuente tener muchas pequeñas ganancias y algunas grandes pérdidas ocasionalmente.


```python
# @title
from scipy.stats import skewnorm
# Crear una figura con 3 subplots alineados
def moda_int(data, bin_v=10):
  counts, bin_edges =np.histogram(data, bins=bin_v)
  modal_index=np.argmax(counts)
  L = bin_edges[modal_index]             # Límite inferior del intervalo modal
  f_m = counts[modal_index]              # Frecuencia del intervalo modal
  f_m_minus_1 = counts[modal_index - 1]  # Frecuencia del intervalo anterior
  f_m_plus_1 = counts[modal_index + 1]   # Frecuencia del intervalo siguiente
  w = bin_edges[modal_index+1] - bin_edges[modal_index]        # Ancho del intervalo
  # Calcular la moda usando la fórmula
  moda = L + ((f_m - f_m_minus_1) / (2*f_m - f_m_minus_1 - f_m_plus_1)) * w
  return moda

fig3, axs = plt.subplots(1, 3, figsize=(18, 6))
bins=20

# Distribución con asimetría negativa
a_neg = 5  # Parámetro de asimetría negativa
data_neg = skewnorm.rvs(-a_neg, size=10000)
mean_neg = np.mean(data_neg)
median_neg = np.median(data_neg)
counts, bin_edges,patches =axs[0].hist(data_neg, bins=bins, color='lightblue', edgecolor='black', alpha=0.7)
mode_neg = moda_int(data_neg, bin_v=bins)
axs[0].axvline(mean_neg, color='red', linestyle='dashed', linewidth=2, label='Media')
axs[0].axvline(median_neg, color='green', linestyle='dashed', linewidth=2, label='Mediana')
axs[0].axvline(mode_neg, color='blue', linestyle='dashed', linewidth=2, label='Moda')
axs[0].set_title('Asimetría Negativa')
axs[0].legend()

# Distribución simétrica (normal)
data_sym = np.random.normal(0, 1, 10000)
mean_sym = np.mean(data_sym)
median_sym = np.median(data_sym)
axs[1].hist(data_sym, bins=bins, color='lightgreen', edgecolor='black', alpha=0.7)
mode_sym = moda_int(data_sym, bin_v=bins)
axs[1].axvline(mean_sym, color='red', linestyle='dashed', linewidth=2, label='Media:')
axs[1].axvline(median_sym, color='green', linestyle='dashed', linewidth=2, label='Mediana')
axs[1].axvline(mode_sym, color='blue', linestyle='dashed', linewidth=2, label='Moda')
axs[1].set_title('Simétrica')
axs[1].legend()

# Distribución con asimetría positiva
a_pos = 5  # Parámetro de asimetría positiva
data_pos = skewnorm.rvs(a_pos, size=10000)
mean_pos = np.mean(data_pos)
median_pos = np.median(data_pos)
mode_pos = moda_int(data_pos, bin_v=bins)

axs[2].hist(data_pos, bins=bins, color='lightcoral', edgecolor='black', alpha=0.7)
axs[2].axvline(mean_pos, color='red', linestyle='dashed', linewidth=2, label='Media')
axs[2].axvline(median_pos, color='green', linestyle='dashed', linewidth=2, label='Mediana')
axs[2].axvline(mode_pos, color='blue', linestyle='dashed', linewidth=2, label='Moda')
axs[2].set_title('Asimetría Positiva')
axs[2].legend()

# Ajustar el layout para evitar superposición
plt.tight_layout()

# Mostrar los gráficos
plt.show()
centergraph()
```

###DISTRIBUCIÓN NORMAL O DE GAUSS
La distribución normal, o de Gauss, es una curva en forma de campana que es simétrica, con sus lados izquierdo y derecho siendo imágenes especulares. Esta distribución se concentra en su centro, con menores probabilidades de eventos extremos.

Es fundamental en el análisis estadístico y financiero por dos razones principales:

- Empírica: Se aproxima bien a muchas distribuciones observadas en la naturaleza y en rendimientos financieros.
- Teórica: Según el teorema del límite central, cuando una variable es afectada por muchas causas independientes, tiende a seguir una distribución normal.

La distribución normal es popular también por su facilidad de uso y tiene características clave como:
- Su distribución está definida por los parámetros media ($μ$) y varianza ($σ^2$)
- La media, mediana y moda son iguales.
- Se describe completamente por su media y varianza (o desviación estándar). Cuando la distribución normal tiene media 0 y desviación estándar 1 se habla de distribución normal estándar.
- Si tienes dos variables aleatorias independientes que siguen una distribución normal, cualquier combinación lineal de ellas también sigue una distribución normal.<Br><Br>

Su función de densidad viene dada por:
$$
f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)
$$

## **Curtosis**

### **Definición:**

La curtosis mide la "agudeza" o "apuntamiento" de la distribución de datos en comparación con una distribución normal. Específicamente, indica cómo se distribuyen los datos en las colas y cerca de la media.

En general la curtosis se compara con la de una distribución normal y nos referimos al exceso de curtosis.

### **Cálculo:**
- La fórmula general para calcular la curtosis en una población es:
  $$ \text{Curtosis} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \bar{x}}{s}\right)^4
  $$
  donde:
  - $ n $ es el número de observaciones.
  - $ x_i $ es cada valor del conjunto de datos.
  - $ \bar{x} $ es la media.
  - $ s $ es la desviación estándar.

En una distribución normal la curtosis es 3, por lo que se suele hablar también del **exceso de curtosis**:

 $$ \text{Exceso de Curtosis} = \frac{1}{n} \sum_{i=1}^{n} \left(\frac{x_i - \bar{x}}{s}\right)^4-3
  $$

### **Tipos de Curtosis:**

1. **Leptocúrtica**:
   - Distribuciones con curtosis positiva (mayor que 3).
   - Tienen colas más largas y delgadas, y un pico más alto y estrecho en el centro.
   - Ejemplo: Distribución de datos donde la mayoría de las observaciones están cerca de la media, pero hay algunos valores extremadamente altos o bajos.

2. **Mesocúrtica**:
   - Distribuciones con curtosis cercana a 3.
   - Tiene una forma similar a la distribución normal.
   - Ejemplo: La distribución normal estándar.

3. **Platicúrtica**:
   - Distribuciones con curtosis negativa (menor que 3).
   - Tienen colas más cortas y gruesas, y un pico más bajo y ancho en el centro.
   - Ejemplo: Distribuciones donde los valores están más dispersos, con menos datos cerca de la media.

- **Curtosis positiva (leptocúrtica)**: Indica una distribución con colas más gruesas que la distribución normal, más probable tener valores extremos.
- **Curtosis negativa (platicúrtica)**: Indica una distribución con colas menos pesadas que la normal, y un pico menos pronunciado.
- **Curtosis cercana a 0**: Indica una distribución similar a la normal.<br><br>


**¿Por qué es relevante en finanzas?** Desde el punto de vista del riesgo. Si los redimientos de un activo presentan una distribución leptocúrtica las probabilidades de tener algún rendimiento muy negativo son más alta.

## Distribución Uniforme
Existen dos tipos principales de distribución uniforme:

- **Distribución Uniforme Discreta:**
   - Esta distribución se aplica a variables discretas (como números enteros).
   - Todos los valores posibles en un rango finito tienen la misma probabilidad.
   - Un ejemplo clásico es lanzar un dado justo, donde cada cara tiene una probabilidad de $\frac{1}{6}$ de salir.

- **Distribución Uniforme Continua:**
   - Se aplica a variables continuas.
   - La probabilidad de que la variable tome un valor dentro de un intervalo específico es uniforme en todos los subintervalos de igual longitud.
   - Un ejemplo común es seleccionar un número aleatorio entre 0 y 1.

### Función de Probabilidad:
Si $X $ es una variable aleatoria con una distribución uniforme discreta que toma valores $ x_1, x_2, \dots, x_n $, entonces la función de probabilidad $P(X = x_i)$ es:
$$P(X = x_i) = \frac{1}{n} \quad \text{para } i = 1, 2, \dots, n$$

### Propiedades:
- **Esperanza (Media):**
  $\mathbb{E}(X) = \frac{n+1}{2}$
  
- **Varianza:**
  $\text{Var}(X) = \frac{(n^2 - 1)}{12}$

## 2. Distribución Uniforme Continua

### Función de Densidad de Probabilidad (PDF):
Si $ X $ es una variable aleatoria con una distribución uniforme continua en el intervalo $ [a, b] $, entonces la función de densidad de probabilidad $ f(x) $ es:
$$
f(x) = \frac{1}{b-a} \quad \text{para } a \leq x \leq b
$$

### Propiedades:
- **Esperanza (Media):**
 $
  \mathbb{E}(X) = \frac{a + b}{2}
  $

- **Varianza:**
  $
  \text{Var}(X) = \frac{(b-a)^2}{12}
  $$

- **Función de Distribución Acumulada (CDF):**
  $$
  F(x) = \frac{x-a}{b-a} \quad \text{para } a \leq x \leq b
  $$


La distribución uniforme es frecuentemente utilizada en simulaciones y generación de números aleatorios.

##DISTRIBUCIÓN BINOMIAL
La distribución binomial es una de las distribuciones de probabilidad discreta más importantes en estadística. Describe el número de éxitos en una secuencia de $n$ experimentos independientes, cada uno de los cuales tiene un resultado binario (éxito o fracaso), con una probabilidad fija $p$ de éxito. Por ejemplo repetir n veces el lanzamiento de una moneda.
**Parámetros:**
- $n$: Número de ensayos (o pruebas) independientes.
- $p$: Probabilidad de éxito en un solo ensayo.
- $q=1-p$: Probabilidad de fracaso en un solo ensayo.
**Función de probabilidad**:
La función de probabilidad de la distribución binomial se define como:
$$P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}$$

  donde $ \binom{n}{k} $ representa el número de maneras en que se pueden obtener $k$ éxitos en $n$ ensayos.
- $X$ puede tomar cualquier valor entero desde 0 hasta $n$ (el número total de ensayos).

**Media (Esperanza matemática)**:
   - La media de la distribución binomial es:
    $
     \mu = np
     $
     Esto representa el número esperado de éxitos en $ n $ ensayos.

**Varianza**:
   - La varianza de la distribución binomial es:
    $
     \sigma^2 = np(1-p)
     $
     Esto mide la dispersión de la distribución alrededor de la media.
   - Para valores grandes de $ n $ y $ p $ no extremadamente cerca de 0 o 1, la distribución binomial puede aproximarse a una distribución normal con media $ np $ y varianza $ np(1-p) $ (teorema del límite central).
   - La distribución binomial es una generalización de la distribución de Bernoulli, que es un caso especial con $ n = 1 $.

Ejemplo práctico:

Supongamos que tiramos una moneda 10 veces ($ n = 10 $) y queremos saber la probabilidad de obtener exactamente 3 caras ($ p = 0.5 $).

La probabilidad de obtener exactamente 3 caras en 10 lanzamientos de moneda se puede calcular como:
$P(X = 3) = \binom{10}{3} (0.5)^3 (0.5)^{7} = \frac{10!}{3!(10-3)!} (0.5)^{10}$

##DISTRIBUCIÓN LOG-NORMAL

La **distribución log-normal** es una distribución de probabilidad continua que describe una variable aleatoria cuyo logaritmo natural sigue una distribución normal. En otras palabras, si una variable aleatoria $X$ tiene una distribución normal, entonces $Y = e^X$ sigue una distribución log-normal.
- Matemáticamente, si $X \sim \mathcal{N}(\mu, \sigma^2)$, entonces $Y = e^X$ sigue una distribución log-normal, denotada como $Y \sim \text{Log-Normal}(\mu, \sigma^2)$.
- **Media**:
$$Media=\exp(μ+\frac{\sigma^2)}{2}$$
- **Varianza**:
   - La varianza de la distribución log-normal es:
     $$
     \text{Var}(Y) = \left(\exp(\sigma^2) - 1\right)̇·\exp\left(2\mu + \sigma^2\right)
     $$

- **Asimetría y Curtosis**:
   - La distribución log-normal es asimétrica hacia la derecha (tiene una cola más larga hacia la derecha).
   - Su coeficiente de asimetría es positivo.
   - La curtosis de la distribución log-normal es mayor que la de una distribución normal, lo que indica colas más pesadas.

6. **Propiedades**:
   - **No Simetría**: A diferencia de la distribución normal, la distribución log-normal no es simétrica.
   - **Producto de Variables Aleatorias**: Si $X_1, X_2, \dots, X_n$ son variables independientes y log-normales, entonces el producto de estas variables $Y = X_1 \cdot X_2 \cdots X_n$ sigue una distribución log-normal.
   - **Relación con la Normalidad**: Como la log-normalidad se define en función de la normalidad, muchas propiedades de la normalidad pueden traducirse a la log-normalidad.

En finanzas distribución log-normal se usa comúnmente para modelar precios de activos  ya que los precios de los activos no pueden ser negativos y suelen crecer de manera exponencial.


## Ejemplo:

Supongamos que el logaritmo natural de una variable aleatoria $Y$ sigue una distribución normal con media $\mu = 0$ y desviación estándar $\sigma = 1$. Entonces, $Y$ sigue una distribución log-normal.

Si quieres encontrar la probabilidad de que $Y$ sea menor que un cierto valor $y_0$, puedes calcular:

$$
P(Y \leq y_0) = P(\log(Y) \leq \log(y_0)) = P(X \leq \log(y_0))
$$

donde $X$ es una variable aleatoria normal con los parámetros dados.

La distribución log-normal es muy útil en situaciones donde los datos se distribuyen de forma no simétrica y positiva, con una cola larga hacia la derecha, lo que es común en muchos fenómenos económicos.


#Covarianza

La **covarianza** es una medida que indica el grado en que dos variables aleatorias varían juntas. Mide cómo cambia una variable en relación con otra. Si las variables tienden a aumentar o disminuir juntas, la covarianza será positiva; si una tiende a aumentar mientras la otra disminuye, la covarianza será negativa. Si no hay una relación lineal clara entre las dos variables, la covarianza será cercana a cero.

### Fórmula de la Covarianza

Para dos variables aleatorias $X$ e $Y$, la covarianza se calcula como:

$$
\text{Cov}(X, Y) = E[(X - E[X])(Y - E[Y])]
$$

En una muestra de datos, la covarianza se calcula como:

$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

donde:
- $X_i$ e $Y_i$ son los valores individuales de las variables $X$ e $Y$.
- $\bar{X}$ y $\bar{Y}$ son las medias de las variables $X$ e $Y$.
- $n$ es el número de observaciones.

- **Covarianza positiva**: Indica que las dos variables tienden a moverse en la misma dirección. Cuando una variable aumenta, la otra también lo hace.
- **Covarianza negativa**: Indica que las dos variables tienden a moverse en direcciones opuestas. Cuando una variable aumenta, la otra tiende a disminuir.
- **Covarianza cercana a cero**: Indica que no hay una relación lineal clara entre las dos variables.

## Correlación

La **correlación** es una medida estadística que indica la fuerza y la dirección de una relación lineal entre dos variables. A diferencia de la covarianza, la correlación está normalizada, lo que significa que siempre se encuentra en el rango de -1 a 1, independientemente de las unidades de medida de las variables.

La correlación entre dos variables $X$ e $Y$ se calcula como:

$$
\rho_{X,Y} = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y}
$$

donde:
- $\text{Cov}(X, Y)$ es la covarianza entre $X$ e $Y$.
- $\sigma_X$ y $\sigma_Y$ son las desviaciones estándar de $X$ e $Y$ respectivamente.

Por tanto si:

- **$\rho = 1$**: Correlación positiva perfecta. Ambas variables se mueven en la misma dirección con una relación lineal perfecta.
- **$\rho = -1$**: Correlación negativa perfecta. Ambas variables se mueven en direcciones opuestas con una relación lineal perfecta.
- **$\rho = 0$**: No hay correlación lineal entre las variables. No significa necesariamente que no haya relación, solo que no es lineal.

## Diferencias Clave entre Covarianza y Correlación

1. **Escala**: La covarianza está afectada por las unidades de las variables, mientras que la correlación es adimensional, es decir, no depende de las unidades de medida.

2. **Interpretación**: La covarianza indica la dirección de la relación (positiva o negativa), pero no proporciona información sobre la fuerza de la relación. La correlación, en cambio, mide tanto la dirección como la fuerza de la relación lineal entre dos variables.

3. **Normalización**: La correlación es esencialmente una covarianza normalizada, lo que permite comparar relaciones lineales entre diferentes pares de variables, incluso si tienen diferentes unidades o escalas.

Destacar 2 elementos relevantes de estas medidas de relación:
- Correlación **no significa causalidad**
- Miden la **relación lineal**. Puede haber variables muy relacionadas no linealmente pero con correlación 0. Por eso hay otras medidad de asociación que miden relaciones entre variables sin asumir que esta es lineal (Spearman, $\tau\text{ de Kendall}$,...)

# Distribución Normal Multivariante

La **distribución normal multivariante** amplía la normal univariante (o gaussiana) a un espacio multidimensional.

La distribución normal multivariante describe el comportamiento de un vector aleatorio de $n$ variables aleatorias, que pueden tener correlaciones entre sí, en lugar de ser independientes. Si $\mathbf{X} = (X_1, X_2, \dots, X_n) $ es un vector aleatorio de $n $ variables, entonces  $ \mathbf{X} $ sigue una distribución normal multivariante si todas las combinaciones lineales de las variables siguen una distribución normal.

La notación común es:

$$\mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma}) $$

donde:
-  $ \boldsymbol{\mu}  $ es el vector de medias (un vector de tamaño  $n$).
-  $ \boldsymbol{\Sigma}  $ es la matriz de covarianzas (una matriz  $n \times n $ que describe las varianzas y las covarianzas entre las variables).

### Función de densidad
La función de densidad de probabilidad (PDF) de la distribución normal multivariante para un vector aleatorio  $\mathbf{X} $ con  $n $ variables es:

$$f(\mathbf{x}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu})^\top \boldsymbol{\Sigma}^{-1} (\mathbf{x} - \boldsymbol{\mu}) \right)$$

Donde:
-  $\mathbf{x} $ es el vector de observaciones.
-  $\boldsymbol{\Sigma} $ es la matriz de covarianza.
-  $|\boldsymbol{\Sigma}| $ es el determinante de la matriz de covarianza.
-  $\boldsymbol{\Sigma}^{-1} $ es la inversa de la matriz de covarianza.


### Propiedades de la Normal Multivariante
- **Distribuciones marginales**: Si $ \mathbf{X} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma}) $, entonces cualquier subconjunto de  $ \mathbf{X}  $ también sigue una distribución normal multivariante.
- **Distribuciones condicionales**: Las distribuciones condicionales de variables en una distribución normal multivariante también son normales. Es decir, la distribución de una variable dada otra también es normal.
-En la distribución normal multivariante, si dos variables aleatorias tienen una covarianza igual a cero, esto no solo significa que son incorreladas, sino que también son **independientes**. Esta es una propiedad particular de la normal multivariante que no se cumple en otras distribuciones.

# Suma de Variables Normales

La **suma de variables normales** sigue una propiedad importante en la teoría de probabilidad y estadística: la suma de dos o más variables aleatorias normales es también una variable aleatoria normal, bajo ciertas condiciones.

Si $X$ y $Y$ son dos variables aleatorias normales independientes:

$$
X \sim \mathcal{N}(\mu_X, \sigma_X^2)
$$
$$
Y \sim \mathcal{N}(\mu_Y, \sigma_Y^2)
$$

Entonces, la suma $Z = X + Y$ también sigue una distribución normal, donde:

$$
Z \sim \mathcal{N}(\mu_X + \mu_Y, \sigma_X^2 + \sigma_Y^2)
$$

Es decir:
- La media de la suma $Z$ es la suma de las medias de $X$ y $Y$: $ \mu_Z = \mu_X + \mu_Y $.
- La varianza de la suma $Z$ es la suma de las varianzas de $X$ y $Y$: $ \sigma_Z^2 = \sigma_X^2 + \sigma_Y^2 $.

### Ejemplo:

Si $X \sim \mathcal{N}(2, 1)$ y $Y \sim \mathcal{N}(3, 4)$, entonces la suma:

$$
Z = X + Y \sim \mathcal{N}(2 + 3, 1 + 4) = \mathcal{N}(5, 5)
$$

En este caso, $Z$ tiene media 5 y varianza 5.

¿Y si tenemos repeticiones de un mismo experimento de manera independiente? Es decir si sumamos n veces la variable $x$?
- $ media = n⋅ \mu_X $.
- $ varianza = n⋅ \sigma_X^2 $
- $ DT = \sqrt n⋅ \sigma_X $


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import multivariate_normal

# Función para generar superficies 3D de las distribuciones normales bivariadas
def plot_bivariate_surface(correlations, mu=[0, 0], sigma=[1, 1], num_samples=1000):
    fig = plt.figure(figsize=(18, 6))

    for i, rho in enumerate(correlations):
        # Crear la matriz de covarianza
        cov = [[sigma[0]**2, rho * sigma[0] * sigma[1]], [rho * sigma[0] * sigma[1], sigma[1]**2]]

        # Definir el grid
        x = np.linspace(-3, 3, 100)
        y = np.linspace(-3, 3, 100)
        X, Y = np.meshgrid(x, y)
        pos = np.dstack((X, Y))

        # Crear la distribución normal bivariada
        rv = multivariate_normal(mu, cov)
        Z = rv.pdf(pos)

        # Crear la gráfica de superficie
        ax = fig.add_subplot(1, 3, i+1, projection='3d')
        ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
        ax.set_title(f"Correlación: {rho}")
        ax.set_xlabel("X1")
        ax.set_ylabel("X2")
        ax.set_zlabel("Densidad")

    plt.tight_layout()
    plt.show()

# Correlaciones de interés: -0.8, 0, 0.8
correlations = [-0.8, 0, 0.8]

# Graficar superficies 3D de las distribuciones normales bivariadas con las correlaciones dadas
plot_bivariate_surface(correlations)
```

# Regresión Lineal

La **regresión lineal** es un método estadístico utilizado para modelar la relación entre una variable dependiente (también llamada variable de respuesta u objetivo) y una o más variables independientes (también conocidas como predictores o explicativas). Cuando solo hay una variable independiente, se denomina **regresión lineal simple**; si hay varias, se llama **regresión lineal múltiple**.

En la regresión lineal simple, se asume que la relación entre la variable independiente $X$ y la variable dependiente $Y$ puede describirse mediante una línea recta. La ecuación general es:

$$
Y = \beta_0 + \beta_1X + \epsilon
$$

donde:
- $Y$ es la variable dependiente.
- $X$ es la variable independiente.
- $\beta_0$ es la intersección con el eje $Y$ (el valor de $Y$ cuando $X = 0$).
- $\beta_1$ es la pendiente de la línea (cuánto cambia $Y$ por cada unidad de cambio en $X$).
- $\epsilon$ es el término de error (la parte de $Y$ que no puede explicarse por la relación lineal con $X$).

Si es mútiple:
$$
Y = \beta_0 + \beta_1X_1 +\beta_2X_2+...+\beta_nX_n \epsilon
$$
donde:
- $X_i$ son las variable independiente.

```python
# @title
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar los datos
np.random.seed(0)
x = np.linspace(-5, 5, 50).reshape(-1, 1)
noise = np.random.normal(0, 4, x.shape)
y = 1 + 2 * x + noise

# Ajustar la regresión lineal
reg = LinearRegression().fit(x, y)
y_pred = reg.predict(x)

# Escoger un punto (el punto número 50)
point_index = 25
x_point = x[point_index]
y_point = y[point_index]
y_pred_point = y_pred[point_index]

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='blue', label='Datos')
plt.plot(x, y_pred, color='red', label=f'Recta de regresión: Y = {reg.intercept_[0]:.2f} + {reg.coef_[0][0]:.2f}X')

# Añadir línea desde el punto a la recta de regresión
plt.plot([x_point, x_point], [y_point, y_pred_point], color='green', linestyle='--')
plt.text(x_point + 0.3, (y_point + y_pred_point) / 2, 'Error', fontsize=12, color='green')

# Marcar el punto específico
plt.scatter(x_point, y_point, color='black')
plt.scatter(x_point, y_pred_point, color='red')

# Configurar etiquetas y leyenda
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ejemplo Recta de Regresión')
plt.xticks([])
plt.yticks([])
plt.show()
```

Los parámetros $\beta_i$ se calcular buscando aquellos valores que minimicen el error cuadrático medio (mean square error) entre los valor de y que hemos observado y los valores que estimamos con nuestra recta de regresión.

El MSE se calcula como:
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^n (Y_i - \hat{Y}_i)^2
$$

donde:
- $Y_i$ son los valores observados de la variable dependiente.
- $\hat{Y}_i$ son los valores predichos por el modelo.
- $n$ es el número total de observaciones.

Por tanto el método de mínimos cuadrados busca los valores de los coeficientes que minimizan:

$$
\text{MSE}(\boldsymbol{\beta}) = \frac{1}{n} \sum_{i=1}^n (Y_i - (\beta_0 + \beta_1X_{1i} + \dots + \beta_kX_{ki}))^2
$$

Para una regresión simple es fácil demostrar que el valor de los coeficientes viene dado por:
$$
\beta_1 = \frac{\text{Cov}(X, Y)}{\text{Var}(X)}
$$

$$
\beta_0 = \bar{Y} - \beta_1 \bar{X}
$$

## MÚLTIPLES VARIABLES. RESULTADOS PARA UNA CARTERA CON *n* ACTIVOS


#### 1. **Media de la Rentabilidad de la Cartera**

La media de la rentabilidad esperada de una cartera, denotada como $R_p$, se calcula como una combinación ponderada de las rentabilidades esperadas de los activos que componen la cartera.

**Fórmula:**

$$
R_p = \sum_{i=1}^{n} w_i \cdot R_i
$$

**Donde:**
- $R_p$: Media de la rentabilidad esperada de la cartera.
- $w_i$: Peso del activo *i* en la cartera.
- $R_i$: Rentabilidad esperada del activo *i*.
- $n$: Número de activos en la cartera.

#### 2. **Varianza de la Rentabilidad de la Cartera**

La varianza de la cartera, denotada como $\sigma_p^2$, mide la volatilidad (o riesgo) de la cartera agregada.

Como ya sabemos la covarianza entre dos activos viene dadad por:
$$
\text{Cov}(R_1, R_2) = \frac{1}{n} \sum_{i=1}^{n} (r_i - \bar{r_1})(r_i - \bar{r_2})
$$

Llamamos **matriz de covarianzas** a la matriz cuadrada que contiene las covarianzas entre los distintos pares de activos de la cartera. Cada elemento de la matriz $\Sigma_{ij}$ representa la covarianza entre el activo *i* y el activo *j*.

$$
\Sigma =
\begin{pmatrix}
\sigma_1^2 & \text{Cov}(R_1, R_2) & \dots & \text{Cov}(R_1, R_n) \\
\text{Cov}(R_2, R_1) & \sigma_2^2 & \dots & \text{Cov}(R_2, R_n) \\
\vdots & \vdots & \ddots & \vdots \\
\text{Cov}(R_n, R_1) & \text{Cov}(R_n, R_2) & \dots & \sigma_n^2
\end{pmatrix}
$$

A partir de la matriz de covarianzas podemos calcular la varianza agregada de nuestra cartera que será
$$
\sigma_p^2 =\sum{\sum{w_i w_j cov_{i,j}}} =\mathbf{w}^T \cdot \Sigma \cdot \mathbf{w}
$$

donde:
- $\sigma_p^2$: Varianza de la cartera.
- $\mathbf{w}$: Vector de pesos de los activos en la cartera, $\mathbf{w}^T = (w_1, w_2, \dots, w_n)$.
- $\Sigma$: Matriz de covarianzas.

Por tanto la desviación típica o **volatilidad** será:

$$
\sigma_p = \sqrt{\sigma_p^2}
$$

Para una cartera con solo dos activos, la varianza se puede simplificar a:

$$
\sigma_p^2 = w_1^2 \cdot \sigma_1^2 + w_2^2 \cdot \sigma_2^2 + 2 \cdot w_1 \cdot w_2 \cdot \text{Cov}(R_1, R_2)
$$ <BR><BR>

¿Y si quiero simular más de un activo?

Existen también técnicas para ello. Introduciremos un modelo básico para simular variables normales con determinada correlación.

Partimos de que ya sabemos simular x variables normales con media $\mu$ y volatilidad $\sigma$. Ahora lo que buscamos es generar esas normales pero manteniendo una determinada estructura de covarianzas/correlaciones $\Sigma$.

Para ello necesitamos introducir el concepto de descomposición de Cholesky.

La **descomposición de Cholesky** es una técnica en álgebra lineal utilizada para descomponer una matriz simétrica y definida positiva en el producto de una matriz triangular inferior y su traspuesta. Es fundamental en muchas aplicaciones numéricas, especialmente en la solución de sistemas de ecuaciones lineales, simulaciones aleatorias, y optimización.

## Definición

Dada una matriz simétrica y definida positiva $ A $, la descomposición de Cholesky produce una matriz triangular inferior $ L $ tal que:

$
A = L L^T
$

Donde:
- $ L $ es una matriz triangular inferior con entradas en las diagonales principales no negativas.
- $ L^T $ es la traspuesta de $ L $.

## Propiedades

1. **Unicidad**: La descomposición de Cholesky es única si la matriz $ A $ es definida positiva y simétrica.

2. **Condiciones**: La matriz $ A $ debe ser simétrica y definida positiva para que exista la descomposición de Cholesky. Esto significa que para todo vector no nulo $ x $, se cumple que $ x^T A x > 0 $.

## Ejemplo

Supongamos que tenemos una matriz simétrica y definida positiva $ A $:

$
A = \begin{pmatrix}
4 & 12 & -16 \\
12 & 37 & -43 \\
-16 & -43 & 98
\end{pmatrix}
$

La descomposición de Cholesky de $ A $ produce una matriz triangular inferior $ L $ tal que:

$
L = \begin{pmatrix}
2 & 0 & 0 \\
6 & 1 & 0 \\
-8 & 5 & 3
\end{pmatrix}
$

Entonces:

$
A = LL^T = \begin{pmatrix}
2 & 0 & 0 \\
6 & 1 & 0 \\
-8 & 5 & 3
\end{pmatrix}
\begin{pmatrix}
2 & 6 & -8 \\
0 & 1 & 5 \\
0 & 0 & 3
\end{pmatrix}
$


 En la generación de muestras de una distribución normal multivariante, la descomposición de Cholesky se utiliza para transformar un conjunto de variables normales estándar en variables normales correlacionadas.


¿Cuál sería la descomposición para una matriz de correlación de dos variables?

$
\text{Matriz Correlaciones} = \begin{pmatrix}
1 & p \\
p & 1\end{pmatrix}=LL^T=\begin{pmatrix}
a & 0 \\
b & c\end{pmatrix}\begin{pmatrix}
a & 0 \\
b & c\end{pmatrix}
$

Resolvemos el sistema de ecuaciones y tenemos:
$$
\text{Matriz Correlaciones} = \begin{pmatrix}
1 & p \\
p & 1\end{pmatrix}=LL^T=\begin{pmatrix}
a & 0 \\
b & c\end{pmatrix}\begin{pmatrix}
a & 0 \\
b & c\end{pmatrix}=\begin{pmatrix}
a^2 & ab \\
ab & b^2+c^2\end{pmatrix}
$$
Y resolvemos igualando término y obteniendo el sistema de ecuaciones. De dónde tenemos que:
$$
\begin{aligned}
a^2 &= 1 \\
ab&=p \\
b^2+c^2&=1
\end{aligned}
⇒\begin{aligned}
a&= 1 \\
b&=p \\
c&=\sqrt{1-ρ^2 }
\end{aligned}⇒L=\begin{pmatrix}
1 & 0 \\
p & \sqrt{1-ρ^2 }\end{pmatrix}
$$

Vamos a usar $L$:
- Primero simulamos $\mathbf{z}=(z_1,z_2$) como normales estándar con medias 0 y desviaciones típicas 1 como ya sabemos:
- Segundo calculamos las variables $\mathbf{y}=(y_1,y_2)$ tales que:

$$
\begin{pmatrix}
y_1 & y_2 \end{pmatrix}=\begin{pmatrix}
z_1&z_2\end{pmatrix}\begin{pmatrix}
1 & 0 \\
p & \sqrt{1-ρ^2 }\end{pmatrix}
$$
Esto nos genera  variables $(Y_1,Y_2) \sim \mathcal{N}(0, 1)$ que tenemos que ajustar para que tengan la media y desviación buscada. Esto es:
$$
\begin{pmatrix}
x_1 & x_2 \end{pmatrix}=\begin{pmatrix}
y_1\sigma_1+\mu_1&y_2\sigma_2+\mu_2\end{pmatrix}
$$

Esto es extensible para n variables, es decir una vez tenemos L, la fórmula general será:
$$
X=LZ^T\Sigma_z+u
$$

```python
#Simulamos 500 paths cadad uno con 250 días:
mu=0.0002
sigma=0.02
escenarios=np.random.normal(mu, sigma,size=(250,500))
#Calculamos las rentabilidades acumuladas
cum_return=np.cumprod((1+escenarios), axis=0)
# Añadir la fila de 1s al inicio de data
cum_return= np.vstack([np.ones((1, cum_return.shape[1])), cum_return])
#Calculamos los paths de precios multiplicando por el precio inicial
path=cum_return*30
```
